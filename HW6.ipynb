```python
import quandl
import matplotlib.pyplot as plt
%matplotlib inline
with open("key.txt","r") as f:
    ke=f.read()
```

```python
armenia=quandl.get('WORLDBANK/ARM_TOT',authtoken=ke)
luxen=quandl.get('WORLDBANK/LUX_TOT',authtoken=ke)
plt.plot(armenia,)
plt.plot(luxen)
plt.show()
```

```python
data=armenia.sort_values(by='Value', ascending=0)
max_2=data['Value'][1]
print '2nd higest value for TOT of Armenia is ' + str(max_2)
```

```python
armenia_imp=quandl.get('WTO/MERCH_IMP_ARM',authtoken=ke)
armenia_exp=quandl.get('WTO/MERCH_EXP_ARM',authtoken=ke)
net_exp=armenia_exp-armenia_imp
plt.plot(net_exp)
plt.show()
```

```python
three_stocks=quandl.get(['WIKI/AAPL','WIKI/AMZN','WIKI/TSLA'],trim_start="2013-07-15",column_index=4, authtoken=ke)
correlation=three_stocks.corr()
print(correlation)
```

```python
import googlemaps
with open("gmaps.txt","r") as f:
    key=f.read()
gmaps=googlemaps.Client(key)
distances=[]
cities=["Helena, MT","Las Vegas, NV","Sacramento, CA",'Seattle, WT',"San Jose, CA",
           "Boston, MA",'Columbus, OH', "Deleware, DE",'Dublin, GE',"New Orleans, LA"]
for i in cities:
    directions=gmaps.directions(i,"New York, NY",units="metric") 
    distances.append(directions[0]['legs'][0]['distance'])
for i in range(0,len(distances)):
    print('The distance between ' +  cities[i] + ' and ' + " New York, NY" + ' is ' + distances[i]["text"])
```
